```python
def invoke(prompt):
    response = client.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content
```

# Pipeline

```python
task = """Heres a very detailed prompt to guide someone in creating a clear and well-structured article about a physics concept:

---

### **Prompt to Write a Physics Concept Article**

#### **Purpose**  
Write an engaging and educational article explaining a specific physics concept to a general audience. The goal is to make the concept accessible, accurate, and interesting, even for those without a strong background in physics.  

---

### **1. Choose a Concept**  
Select a physics concept that:  
- Is well-defined and can be explained clearly. Examples include **Newton's Laws of Motion, Quantum Tunneling, Relativity, Entropy, The Doppler Effect**, etc.  
- Has practical applications or real-world examples.  
- Can be simplified for an audience unfamiliar with technical jargon.

**Prompt Examples:**  
- *"Explain Einstein's Theory of Special Relativity and how time dilation works."*  
- *"Describe Newton's 3 Laws of Motion with real-life examples like sports or driving a car."*  

---

### **2. Audience**  
Define your audience clearly:  
- A general reader (high school students, non-science readers).  
- Readers with basic science curiosity but no formal physics training.

---

### **3. Structure of the Article**  

#### **Introduction (Hook + Overview)**  
- **Start with a hook**: Begin with an intriguing fact, a question, or a real-life scenario related to the concept.  
  - Example: *"Have you ever wondered why astronauts float in space while youre firmly planted on Earth?"*  
- **Introduce the concept**: Name the physics principle and briefly explain its importance or relevance.  
- **Purpose statement**: Tell the reader what theyll learn and why it matters.  

---

#### **Section 1: Background/Context**  
- Briefly explain the historical or scientific background of the concept.  
- Include any significant discoveries, the scientists involved, or why it was groundbreaking.  
  - Example: *"In 1687, Sir Isaac Newton introduced the laws of motion, transforming how we understand force and movement."*  

---

#### **Section 2: The Core Explanation**  
- Break the concept into **clear, logical steps**. Use simple language, avoiding jargon unless defined.  
- **Key requirements for this section**:  
   - Use analogies or comparisons to everyday objects and experiences.  
     - Example: *"Entropy can be compared to a messy room  over time, it naturally becomes disorganized unless you put energy into cleaning it."*  
   - Include diagrams, thought experiments, or hypothetical examples to make abstract ideas more concrete.  
     - Example: *Imagine two twins: one stays on Earth, while the other travels in a spaceship near the speed of light. When the space traveler returns, theyve aged less.*  
   - Explain any formulas or equations step-by-step with context (if applicable). Use only basic math if possible.  

---

#### **Section 3: Real-Life Applications**  
- Explain where and how the concept applies in the real world.  
- Examples:  
  - *"Quantum tunneling is essential for technologies like flash memory in your computer."*  
  - *"The Doppler effect explains why ambulance sirens sound higher when approaching and lower when passing by."*  

---

#### **Section 4: Common Misconceptions**  
- Address any misunderstandings people have about the concept.  
- Example: *"Many people think gravity doesnt exist in space, but astronauts float because they are in free fall."*  

---

#### **Conclusion (Summary + Takeaway)**  
- Summarize the main points of the article in 12 sentences.  
- Leave readers with a thought-provoking question or a final, memorable idea.  
  - Example: *"Next time you throw a ball or see a satellite orbiting Earth, remember that Newtons laws are at play, shaping everything from sports to space exploration."*  

---

### **4. Writing Style**  
- Use **simple and clear language**. Avoid complex sentences.  
- Be **engaging and conversational**. Write as if youre explaining to a curious friend.  
- Add energy through active voice and varied sentence structure.  
- Use short paragraphs and bullet points for clarity.  

---

### **5. Length**  
Target **8001,200 words**. Adjust as necessary based on the complexity of the concept.

---

### **6. Visual Aids and Examples**  
- Suggest including **diagrams, illustrations, or charts** where possible to aid understanding.  
- Use real-world examples, such as sports, technology, or nature, to make the content relatable.  

---

### **Optional Add-Ons**  
- Include a brief Q&A section answering common questions.  
- Provide links or references for further reading.  
- Suggest an experiment at home if possible (e.g., simple demonstrations like measuring the speed of sound with echoes).  

---

By following this structure and style guide, the article will be informative, engaging, and approachable for a broad audience. Let me know if you need further examples or refinements!"""
```

```python
everything_together = """Generate information about laws of physics. Write concisely in a educative manner. Use the summary to write a blog post according to the following conastraints: (1) 500 words, (2) instructive style. validate the law of gravity is discussed"""

#broken down into different parts
generate_information = """Generate information about laws of physics."""
writing_style = """Write concisely in a educative manner."""
write_blogpost = """Use the summary to write a blog post according to the following conastraints: (1) 500 words, (2) instructive style""""
validate_main_points = """validate the law of gravity is discussed"""

invoke( generate_information | writing_style | write_blogpost | validate_main_points)

writing_style = """"

### **4. Writing Style**  
- Use **simple and clear language**. Avoid complex sentences.  
- Be **engaging and conversational**. Write as if youre explaining to a curious friend.  
- Add energy through active voice and varied sentence structure.  
- Use short paragraphs and bullet points for clarity.  

"""

invoke( generate_information | writing_style | write_blogpost | validate_main_points)
```
